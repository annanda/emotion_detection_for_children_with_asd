## Confusion Matrix Using heatmaps for better visualisation

```python
confusion_matrix = np.array([[904, 158, 21, 58], [242, 3003, 51, 215], [8, 19, 80, 16], [76, 158, 18, 658]])
confusion_matrix.shape
```

```python
classes_once = ['Blue', 'Green', 'Red', 'Yellow']

ax = sns.heatmap(confusion_matrix, 
                 annot=True, 
                 fmt=".0f", 
                 xticklabels=classes_once, 
                 yticklabels=classes_once, 
                cmap="binary")
plt.ylabel('True Classes')
plt.xlabel('Predicted Classes')
# ax.xaxis.tick_top()
```
