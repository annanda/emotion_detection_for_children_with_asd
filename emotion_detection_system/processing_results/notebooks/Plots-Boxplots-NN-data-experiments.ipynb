```python
def apply_model_subset_type_to_data(df_line):
    model_subset_type = None
    slug = df_line['Data_Included_Slug'].lower()
    annotation = df_line['Annotation_Type'].lower()
    if 'all_data' in slug:
        model_subset_type = 'GM'
    elif 'participant' in slug:
        model_subset_type = 'SM'
    elif 'session' in slug:
        model_subset_type = 'HSM'
    
    postfixed = model_subset_type
    fusion_type = None
    if 'va_early_fusion' in slug:
        fusion_type = 'EF'
    elif 'va_late_fusion' in slug:
        fusion_type = 'LF'
        
    if fusion_type is not None:
        postfixed = f'{model_subset_type}_{fusion_type}'
        
    postfixed_annotation = model_subset_type
    annotation_type = None
    if 'parents' in annotation:
        annotation_type = 'P'
    elif 'specialist' in annotation:
        annotation_type = 'S'
        
    if annotation_type is not None:
        postfixed_annotation = f'{model_subset_type}_{annotation_type}'
        
    df_line['model_subset_type'] = model_subset_type
    df_line['postfixed_model_subset_type'] = postfixed
    df_line['postfixed_annotation'] = postfixed_annotation
    return df_line


def prefix_baseline_model_subset_type_to_data(df_line):
    df_line['model_subset_type'] = f"BL_{df_line['model_subset_type']}"
    return df_line
    
    
def prepare_comparison_df_clean(dfs_to_append, conf_id_func, fields):
    """
    merge together all dfs_to_append, and apply conf_id_func(df_line) to the field conf_group_id
    filters out so that only the fields are used. "eg: ['Accuracy', 'Accuracy_Balanced']"
    final DF will contain the fields:
        model_subset_type
        postfixed_model_subset_type
        conf_group_id
        [+ fields]
    """
    def _apply_conf_id_to_data(df_line):
        df_line['conf_group_id'] = conf_id_func(df_line)
        return df_line
    
    final_df = None
    extended_fields = fields + ['model_subset_type', 'postfixed_model_subset_type', 'postfixed_annotation']
    for model_group_df in dfs_to_append:
        clean_df = model_group_df[extended_fields]
        clean_df = clean_df.apply(_apply_conf_id_to_data, axis=1)

        if final_df is None:
            final_df = clean_df
        else:
            final_df = final_df.append(clean_df, ignore_index=True)
    return final_df
    
```

```python
def print_metric_boxplot(plot_df, model_group_ids, model_group_field, metric_field, domain, title, show_legend=True):
    "method for printing boxplot"
    model_group_legend='Model Group'
    
    conf_group_field='conf_group_id'
    conf_group_legend='Legend'
    
    if metric_field == 'Accuracy_Balanced':
        metric_title = 'Balanced Accuracy'
    else:
        metric_title = metric_field.replace('_', ' ').capitalize()
    

    final = alt.Chart(
        plot_df
    ).mark_boxplot(
        size=45, 
        outliers=True, 
        median=alt.MarkConfig(color='#ffffff'), 
        ticks=True,
    ).encode(
        x=alt.X(
            f'{model_group_field}:N', 
            sort=model_group_ids,
            title=f'{model_group_legend}'
        ),
        y=alt.Y(
            f'{metric_field}:Q',
            title=f'{metric_title}',
            scale=alt.Scale(
                domain=domain
            )
        ),
        color=alt.Color(
            f'{conf_group_field}:N', title=f'{conf_group_legend}'
        )
    ).properties(
        width=300, height=300, title=title
    ).configure_legend(
        disable=not show_legend,
        orient='bottom',
        titleFontSize=15,
        labelFontSize=15,
        columns=2
    ).configure_axis(
        labelFontSize=15
    )

    final.display()
```

```python
def print_boxplots_for_baseline_vs_overSampling(domain, show_title=True, show_legend=True):
#     if baseline? add all scenarios then
    scenario = ['va_late_fusion', 'va_early_fusion']
    annotation_type = ['parents', 'specialist']
    sub_set_data_list = ['all_data', 'participants', 'sessions']
    fields = ['Accuracy', 'Accuracy_Balanced']
    # in what order should the boxplots displayed
    sort_model_subset_type = ['GM','BL_GM', 'SM','BL_SM', 'HSM','BL_HSM']

    # exp configuration group, this function is applied to each line in the dataframes addin collum 'conf_group_id'
    conf_id_func = lambda df_line: 'Baseline' if 'BL' in df_line['model_subset_type'] else 'OverSampling'
    # comparing the late vs early fusion should change this to the 'postfixed_model_subset_type' field 
    # since it will contain the model_subset_type + 'EF' or 'LF'.    
    model_group_field = 'model_subset_type'
    dfs_to_append = []
    for subset_data in sub_set_data_list:
        resulting_df, baseline_df = get_subset_data(oversampling_data_results, 
                                                scenario=scenario,
                                                annotation_type=annotation_type, 
                                                subdataset_case=subset_data)
         
        # apply model subset field (eg: GM for all_data, etc..) in new field "model_subset_type"
        # also creates "postfixed_model_subset_type" field with the "LF" or "EF" added to the model_subset_type
        resulting_df = resulting_df.apply(apply_model_subset_type_to_data, axis=1)
        
        # apply model subset field to baseline, and then change it to have the prefix for BL 
        baseline_df = baseline_df.apply(
            apply_model_subset_type_to_data, axis=1
        ).apply(
            prefix_baseline_model_subset_type_to_data, axis=1
        )
    
    
        #     append to the DFS list result and baseline    
        dfs_to_append.append(resulting_df)
        # if want to compare to baseline, otherwise, dont append it
        dfs_to_append.append(baseline_df)
    

    #  final_df is the concat DFs from the dfs_to_append 
    final_df = prepare_comparison_df_clean(
        dfs_to_append=dfs_to_append,
        conf_id_func=conf_id_func,
        fields=fields
    )
    
    # for each field (accuyracy or balanced acc) will print the boxplot of that metric
    for field in fields:
        title = ''
        if show_title:
            title = f'{field} for Baseline Vs OverSampling'
        print_metric_boxplot(
            final_df, sort_model_subset_type, model_group_field, field, domain, title, show_legend
        )
```
